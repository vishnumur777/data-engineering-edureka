# Python Decorators

```python
mylist = [1,2,10, 100, -1]
```

```python
def operfunc(func):
    def wrapper(l):
        print([].append(l))
        print(len(l))
        func(l)
        print(tuple(l))
        print(set(l))
    return wrapper
```

```python
@operfunc
def mainl(l):
    l.sort()
    print(l)
```

```python
mainl(mylist)
```

```python
mylist.sort
```

# `yield` Keyword

```python
def fun_num(num):
    result=[]
    for i in num:
        result.append(i*i)
    return result
```

```python
def fun1_num(num):
    result=[]
    for x in num:
        yield x*x
```

```python
a=[1,2,3,4,5]
```

```python
f1 = fun_num(a)
f2 = fun1_num(a)
```

```python
f1
```

```python
f2
```

```python
for x in f2:
    print(x)
```

### `pd.join()`

## Types of Join
- ## Cross join
    - `E x D` (basic)
- ## Inner join
    - cross join + condition=(matching)
- ## Outer join
    - cross join + condition
        - left
        - right
        - full
- ## semi join
    - joins based on one condition
- ## anti join
    - cross join + condition = unmatch

```python
pd.merge(emp, dept, how="cross")
```

```python
pd.merge(emp, dept, how="inner")
```

```python
pd.merge(emp, dept, how="outer")
```

```python
pd.merge(emp, dept, how="left")
```

```python
pd.merge(emp, dept, how="right")
```

```python
pd.merge(emp, dept, on="department_id", how="inner")
```
