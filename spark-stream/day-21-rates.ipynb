```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.master("local[5]").appName("VarunSpark").getOrCreate()
```

```python
df = spark.readStream.format('rate').option('rowsPerSecond',10).load()
```

```python
df.printSchema()
```

```python
df.writeStream.format('console').start().awaitTermination()
```

```python
spark.conf.set('spark.sql.streaming.checkpointLocation','assets/checkpoint')
```

```python
df.writeStream.format('json').option('path','checkpoints_varun').start().awaitTermination()
```

```python
from pyspark.sql.types import * 
```

```python
sch = StructType([ StructField('timestamp',TimestampType(),False),StructField('value',LongType(),False)])
```

```python
df=spark.readStream.format('json').schema(sch).load(r'checkpoints_varun')
```

```python
df.printSchema()
```

```python
res=df.withColumnRenamed('timestamp','ingestionTime')
```

```python
res.printSchema()
```

```python
res.createOrReplaceTempView('timings')
```

```python
resdf=spark.sql('select ingestionTime,value,value+100 as age from timings')
```

```python
resdf.printSchema()
```

```python
resdf.writeStream.format('console').start().awaitTermination()
```
