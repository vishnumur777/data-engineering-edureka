````python
from pyspark import SparkConf, SparkContext
````

````python
sconf = SparkConf().setAppName('webloganalyzev').setMaster('local[3]')
````

````python
sc = SparkContext(conf=sconf)
````

````python
sc
````

````python
logs = sc.textFile('pysparkasset/weblog.log')
````

````python
logs20 = sc.textFile('pysparkasset/weblog.log',20)
````

````python
logs.getNumPartitions()
````

````python
logs.take(4)
````

# Number of request raised by every user?

````python
import re
````

````python
get_ip = logs.flatMap(lambda x: re.findall(r'\d+\.\d+\.\d+\.\d+',x))
````

````python
user_count = get_ip.map(lambda x: (x, 1))
````

````python
user_count = user_count.reduceByKey(lambda x,y: x+y)
````

````python
user_count.sortBy(lambda x: x[1], ascending=False).collect()
````

# Total number of successful response?

````python
sp_logs = logs.map(lambda x: x.split())
````

````python
sp_200_logs = sp_logs.filter(lambda x: x[8]=='200')
````

````python
sp_200_logs.count()
````
