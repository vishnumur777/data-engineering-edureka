```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.master("local[5]").appName("VarunSpark").getOrCreate()
```

```python
df = spark.read.csv("timeseries.csv", sep=",", header=True, inferSchema=True)
```

```python
df.show(5)
```

```python
new_df = df.limit(20).toPandas()
```

```python
new_df.plot()
```

```python
new_df.plot.bar()
```

```python
new_df.plot.barh()
```

```python
new_df.plot.scatter(x="Date", y="Open")
```

```python
new_df.plot.box()
```

```python
new_df.plot.area()
```

```python
new_df.plot.hist()
```
